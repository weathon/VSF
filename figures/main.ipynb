```python
import numpy as np
Is = np.arange(0.5, 3, 0.5)
Js = np.arange(0, 0.6, 0.2)
cols, rows = len(Is), len(Js)

images = [[iio.imread(f'pngs/demo_{i}_{j}.png') for i in range(cols)] for j in range(rows)]

fig, axes = pylab.subplots(rows, cols, figsize=(cols*2, rows*2))
print(axes.shape)
for i in range(rows):
    for j in range(cols):
        axes[i, j].imshow(images[i][j])
        axes[i, j].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
        for spine in axes[i, j].spines.values():
            spine.set_visible(False)
        if i == 0:
            axes[i, j].set_title(f'alpha={Is[j]:.1f}', fontsize=10)
        if j == 0:
            axes[i, j].set_ylabel(f'beta={Js[i]:.1f}', fontsize=10, labelpad=10)

pylab.tight_layout()
pylab.show()
```
